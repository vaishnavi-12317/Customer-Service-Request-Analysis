### 2.2 Missing Value Treatment

#### Remove Closed Date null

```python
df = df[df['Closed Date'].notnull()]
```

```python
print("After removing null values:", df.shape)
df.head()
```

### 2.3 Date Analysis

```python
df['Created Date'] = pd.to_datetime(df['Created Date'], format='%m/%d/%Y %I:%M:%S %p')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], format='%m/%d/%Y %I:%M:%S %p')
```

```python
df[['Created Date','Closed Date']].head()
df.dtypes
```

### 2.3.1 Time Difference

```python
df['Request_Closing_Time'] = df['Closed Date'] - df['Created Date']
```

```python
df[['Created Date','Closed Date','Request_Closing_Time']].head()
```

### 2.3.2 Convert to seconds

```python
df['Request_Closing_Time'] = df['Request_Closing_Time'].dt.total_seconds()
```

### 4.2 Average Closing Time

### Step 5: Response Time Analysis

```python
df.groupby('Complaint Type')['Request_Closing_Time'].mean()
```

### Step 6 & 7: Statistical Test (Kruskal-Wallis)

```python
from scipy.stats import kruskal

groups = []
for name, group in df.groupby('Complaint Type'):
    groups.append(group['Request_Closing_Time'].dropna())

stat, p = kruskal(*groups)

print("P-value:", p)

if p < 0.05:
    print("Reject H0 (Different response times)")
else:
    print("Fail to Reject H0 (Same response times)")
```
